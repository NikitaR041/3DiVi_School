```python
"""
используя библиотеку torch, сравните время выполнения
перемножения матрицы на CPU и GPU (создать матрицу можно так:
torch.randn((height, width)); переместить матрицу на GPU можно
так: torch.randn((height, width)).cuda()). Проверьте с помощью
nvidia-smi, что матрица была перемещена на GPU (была занята
видеопамять).
"""
import torch

torch.cuda.is_available()

height, width = 3000, 3000

A_cpu = torch.randn((height, width))
B_cpu = torch.randn((height, width))

%timeit A_cpu @ B_cpu  # оператор @ — это матричное умножение
```

```python
"""
попробовать инициализировать другой объект в конструкторе и
вызвать один из его методов в методе своего класса;
"""
class Animal:
  def __init__(self, s, x, y):
    self.s = s
    self.x = x
    self.y = y
    print(f"Появился на свет {s}, он расположен на коорд {x}, {y}")

  def sayAnimal(self, s):
    print(f"Ты зачем меня вызвал {s}")

MyObj = Animal("Собкаен", 1, 1)
MyObj.sayAnimal("ГавГАВ")
```
